<h1>Data Transformations

```python
import pandas as pd
```

```python
domy_2010_2015 = pd.read_csv('../Datasets/Average_Prices/RAW_data/DBW_DB_(2010-2015).csv', sep=';')
domy_2010_2015.head()
```

```python

domy_2010_2015_wybrane_kolumny = domy_2010_2015[['jednostka_terytorialna', 'nazwa_pozycja_2', 'id_daty', 'wartosc']]
domy_2010_2015_wybrane_kolumny.head()
```

```python

domy_2010_2015_wybrane_kolumny.columns = ['miasto', 'rynek', 'rok', 'cena']
domy_2010_2015_wybrane_kolumny.head()
```

```python
domy_2010_2015_wybrane_kolumny.info()
```

```python

domy_2010_2015_wybrane_kolumny['rynek'] = domy_2010_2015_wybrane_kolumny['rynek'].replace({
    'pierwotny rynek nieruchomości': 'pierwotny_rynek',
    'wtórny rynek nieruchomości': 'wtorny_rynek',
    'Ogółem': 'laczny_rynek'
})
```

```python
domy_2010_2015_wybrane_kolumny.head()
```

```python

domy_2010_2015_zgrupowane = domy_2010_2015_wybrane_kolumny.copy()
```

```python

domy_2010_2015_zgrupowane = domy_2010_2015_zgrupowane.groupby(['miasto', 'rok', 'rynek'])['cena'].mean()
```

```python

domy_2010_2015_zgrupowane = domy_2010_2015_zgrupowane.unstack(level=['miasto', 'rynek'])
```

```python
domy_2010_2015_zgrupowane.head()
```

```python

domy_2010_2015_splaszczone = domy_2010_2015_zgrupowane.copy()
```

```python

domy_2010_2015_splaszczone.columns = ['_'.join(col).strip() for col in domy_2010_2015_splaszczone.columns.values]
domy_2010_2015_splaszczone.head()
```

```python
domy_2010_2015_splaszczone = domy_2010_2015_splaszczone.reset_index()
```

```python
domy_2010_2015_splaszczone.head(10)
```

```python

domy_2010_2015_splaszczone.to_csv('../Datasets/Average_Prices/Ready _to_use/domy_2010_2015_przerobione.csv', index=False)
```
