```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('../Datasets/Average_Prices/Ready _to_use/domy_2010_2015_przerobione.csv')
df2 = pd.read_csv('../Datasets/Average_Prices/Ready _to_use/domy_2016_2022_przerobione.csv')
df3 = pd.read_csv('../Datasets/Average_Prices/Ready _to_use/domy_2023_przerobione.csv')
```

```python
df3.head()
```

```python
df_all_2010_2022 = pd.concat([df1,df2], ignore_index=True)
df_all_2010_2022.head(70000)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df_all_2010_2022['rok'], df_all_2010_2022['Gdańsk_laczny_rynek'], marker='o', label='Gdańsk')
plt.plot(df_all_2010_2022['rok'], df_all_2010_2022['Kraków_laczny_rynek'], marker='o', label='Kraków')
plt.plot(df_all_2010_2022['rok'], df_all_2010_2022['Poznań_laczny_rynek'], marker='o', label='Poznań')
plt.plot(df_all_2010_2022['rok'], df_all_2010_2022['Warszawa_laczny_rynek'], marker='o', label='Warszaw')
plt.plot(df_all_2010_2022['rok'], df_all_2010_2022['Wrocław_laczny_rynek'], marker='o', label='Wrocław')

plt.xticks(np.arange(2010, 2023, 1))
plt.title('Ceny w latach 2010-2015')
plt.xlabel('Lata')
plt.ylabel('Ceny za m2')

plt.legend()

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))

plt.plot(df3['rok_kwartal'], df3['Gdańsk_laczny_rynek'], marker='o', label='Gdańsk')
plt.plot(df3['rok_kwartal'], df3['Kraków_laczny_rynek'], marker='o', label='Kraków')
plt.plot(df3['rok_kwartal'], df3['Poznań_laczny_rynek'], marker='o', label='Poznań')
plt.plot(df3['rok_kwartal'], df3['Warszawa_laczny_rynek'], marker='o', label='Warszaw')
plt.plot(df3['rok_kwartal'], df3['Wrocław_laczny_rynek'], marker='o', label='Wrocław')

plt.title('Ceny w 2023')
plt.xlabel('Lata')
plt.ylabel('Ceny za m2')

plt.legend()

plt.grid(True)
plt.show()
```
