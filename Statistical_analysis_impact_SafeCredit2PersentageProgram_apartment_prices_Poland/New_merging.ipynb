```python
def mediana_wp(df,miasto):
    #return round(df[(df['date'] == okres) & (df['ownership'] == typ_wlasnosci)]['city'].median()  / df[(df['date'] == okres) & (df['ownership'] == typ_wlasnosci)]['squareMeters'].median(),3)

    return( round(df[(df['city'] == miasto) & (df['date'] == '2023-08') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3),
            round(df[(df['city'] == miasto) & (df['date'] == '2023-09') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3),
            round(df[(df['city'] == miasto) & (df['date'] == '2023-10') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3),
            round(df[(df['city'] == miasto) & (df['date'] == '2023-11') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3),
            round(df[(df['city'] == miasto) & (df['date'] == '2023-12') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3),
            round(df[(df['city'] == miasto) & (df['date'] == '2024-01') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3),
            round(df[(df['city'] == miasto) & (df['date'] == '2024-02') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3),
            round(df[(df['city'] == miasto) & (df['date'] == '2024-03') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3),
            round(df[(df['city'] == miasto) & (df['date'] == '2024-04') & (df['ownership'] == 'condominium')]['price_per_m2'].median(),3)
        )
    
def mediana_ws(df,miasto):
#return round(df[(df['date'] == okres) & (df['ownership'] == typ_wlasnosci)]['city'].median()  / df[(df['date'] == okres) & (df['ownership'] == typ_wlasnosci)]['squareMeters'].median(),3)

    return( round(df[(df['city'] == miasto) & (df['date'] == '2023-08') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3),
        round(df[(df['city'] == miasto) & (df['date'] == '2023-09') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3),
        round(df[(df['city'] == miasto) & (df['date'] == '2023-10') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3),
        round(df[(df['city'] == miasto) & (df['date'] == '2023-11') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3),
        round(df[(df['city'] == miasto) & (df['date'] == '2023-12') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3),
        round(df[(df['city'] == miasto) & (df['date'] == '2024-01') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3),
        round(df[(df['city'] == miasto) & (df['date'] == '2024-02') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3),
        round(df[(df['city'] == miasto) & (df['date'] == '2024-03') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3),
        round(df[(df['city'] == miasto) & (df['date'] == '2024-04') & (df['ownership'] == 'cooperative')]['price_per_m2'].median(),3)
    )
    
```

```python
mediana_df = pd.DataFrame()
mediana_df['okres'] = ['sierpien_2023', 'wrzesien_2023', 'pazdziernik_2023', 'listopad_2023', 'grudzien_2023', 
              'styczen_2024', 'luty_2024', 'marzec_2024', 'kwiecien_2024']


mediana_df['mediana_gdansk_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'gdansk')
mediana_df['mediana_gdansk_wspoldzielnia'] = mediana_ws(domy_credyt_all_v2_wybrane_kolumne_miasta,'gdansk')


mediana_df['mediana_krakow_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'krakow')
mediana_df['mediana_krakow_wspoldzielnia'] = mediana_ws(domy_credyt_all_v2_wybrane_kolumne_miasta,'krakow')


mediana_df['mediana_poznan_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'poznan')
mediana_df['mediana_poznan_wspoldzielnia'] = mediana_ws(domy_credyt_all_v2_wybrane_kolumne_miasta,'poznan')

mediana_df['mediana_warszawa_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'warszawa')
mediana_df['mediana_warszawa_wspoldzielnia'] = mediana_ws(domy_credyt_all_v2_wybrane_kolumne_miasta,'warszawa')

mediana_df['mediana_wroclaw_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'wroclaw')
mediana_df['mediana_wroclaw_wlasnosc_prywatna'] = mediana_ws(domy_credyt_all_v2_wybrane_kolumne_miasta,'wroclaw')



mediana_df.head(10000)
```

```python
mediana_df['mediana_gdansk_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'gdansk')
mediana_df['mediana_krakow_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'krakow')
mediana_df['mediana_poznan_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'poznan')
mediana_df['mediana_warszawa_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'warszawa')
mediana_df['mediana_wroclaw_wlasnosc_prywatna'] = mediana_wp(domy_credyt_all_v2_wybrane_kolumne_miasta,'wroclaw')

```

```python
plt.figure(figsize=(20, 10))

plt.plot(mediana_df['okres'], mediana_df['mediana_gdansk_wlasnosc_prywatna'], marker='o', label='Gdańsk')
plt.plot(mediana_df['okres'], mediana_df['mediana_krakow_wlasnosc_prywatna'], marker='o', label='Kraków')
plt.plot(mediana_df['okres'], mediana_df['mediana_poznan_wlasnosc_prywatna'], marker='o', label='Poznań')
plt.plot(mediana_df['okres'], mediana_df['mediana_warszawa_wlasnosc_prywatna'], marker='o', label='Warszaw')
plt.plot(mediana_df['okres'], mediana_df['mediana_wroclaw_wlasnosc_prywatna'], marker='o', label='Wrocław')



# Dodaj tytuł i etykiety osi
#plt.xticks(np.arange(2010, 2023, 1))
plt.yticks(np.arange(8000, 20000, 1000))
plt.title('Ceny w latach 2010-2015')

plt.xlabel('Lata')
plt.ylabel('Ceny za m2')

# Dodaj legendę
plt.legend()

# Wyświetl wykres
plt.grid(True)
plt.show()
```
