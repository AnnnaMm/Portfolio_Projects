<h2>mediany<h2>

```python
domy_credyt_mediany = domy_credyt_all_filtered = pd.read_csv('../Datasets/Price_During_Credit(08.2023-03.2024)/Ready_to_use/domy_credyt_mediany_wp_ws.csv')
domy_credyt_mediany.head()
```

```python
domy_credyt_mediany_wp = domy_credyt_mediany[['okres','mediana_gdansk_wlasnosc_prywatna','mediana_poznan_wlasnosc_prywatna','mediana_poznan_wlasnosc_prywatna','mediana_warszawa_wlasnosc_prywatna','mediana_wroclaw_wlasnosc_prywatna']]
domy_credyt_mediany_wp.head()
```

```python
domy_credyt_mediany_wp_melt = domy_credyt_mediany_wp.melt(id_vars=["okres"], var_name="Miasto", value_name="wlasnosc_prywatna")

# Definition of a mixed-effects model
model = mixedlm("wlasnosc_prywatna ~ C(okres,Treatment(reference='sierpien_2023'))", domy_credyt_mediany_wp_melt, groups=domy_credyt_mediany_wp_melt["Miasto"])
result = model.fit()

print(result.summary())
```

```python

domy_credyt_mediany_wp_melt = domy_credyt_mediany_wp.melt(id_vars=["okres"], var_name="Miasto", value_name="wlasnosc_prywatna")
domy_credyt_mediany_wp_melt['okres'] = pd.Categorical(domy_credyt_mediany_wp_melt['okres'],
                                                      categories=["sierpien_2023", "wrzesien_2023", "pazdziernik_2023", "listopad_2023","grudzien_2023","styczen_2024","luty_2024","marzec_2024","kwiecien_2024"],
                                                      ordered=True)

# Definition of a mixed-effects model
model = MixedLM.from_formula("wlasnosc_prywatna ~ C(okres, Treatment(reference='sierpien_2023'))", 
                             groups=domy_credyt_mediany_wp_melt["Miasto"], 
                             data=domy_credyt_mediany_wp_melt)
result = model.fit()

print(result.summary())
```

```python
# Melting data
domy_credyt_mediany_wp_melt = domy_credyt_mediany_wp.melt(id_vars=["okres"], var_name="Miasto", value_name="wlasnosc_prywatna")
domy_credyt_mediany_wp_melt['okres'] = pd.Categorical(domy_credyt_mediany_wp_melt['okres'],
                                                      categories=["sierpien_2023", "wrzesien_2023", "pazdziernik_2023", "listopad_2023","grudzien_2023","styczen_2024","luty_2024","marzec_2024","kwiecien_2024"],
                                                      ordered=True)

# Definition of a mixed-effects model
model = MixedLM.from_formula("wlasnosc_prywatna ~ C(okres, Treatment(reference='sierpien_2023'))", 
                             groups=domy_credyt_mediany_wp_melt["Miasto"], 
                             data=domy_credyt_mediany_wp_melt)
result = model.fit()

summary = result.summary()

print(summary)

# # Display only the column of coefficients (Coef.) in percentage format
# coef_table = summary.tables[1]['Coef.']
# coef_table_numeric = pd.to_numeric(coef_table, errors='coerce') * 100
# coef_table_formatted = coef_table_numeric.apply(lambda x: '{:.2f}%'.format(x))
# print(coef_table_formatted)
```

<h1>2010 - 2023<h1>

```python
df1 = pd.read_csv('../Datasets/Average_Prices/Ready _to_use/domy_2010_2015_przerobione.csv')
df2 = pd.read_csv('../Datasets/Average_Prices/Ready _to_use/domy_2016_2022_przerobione.csv')
df3 = pd.read_csv('../Datasets/Average_Prices/Ready _to_use/domy_2023_przerobione_srednie.csv')
df_all_2010_2023 = pd.concat([df1,df2,df3], ignore_index=True)
df_all_2010_2023.head(70000)
```

```python
df_all_2010_2023_wybrane_kolumy_miasta = df_all_2010_2023[['rok','Gdańsk_laczny_rynek','Kraków_laczny_rynek','Poznań_laczny_rynek','Warszawa_laczny_rynek','Wrocław_laczny_rynek']]  
df_all_2010_2023_wybrane_kolumy_miasta.head(10000)
df_all_2010_2023_wybrane_kolumy_miasta.head()
 
```

```python
df_2023_wybrane_kolumy_miasta_copy = df_all_2010_2023_wybrane_kolumy_miasta.copy()
df_2023_wybrane_kolumy_miasta_copy.head()
```

```python
# Melt the dataframe for repeated measures analysis
df_melt = pd.melt(df_2023_wybrane_kolumy_miasta_copy, id_vars=['rok'], value_vars=[
                  'Gdańsk_laczny_rynek', 'Kraków_laczny_rynek', 'Poznań_laczny_rynek', 'Warszawa_laczny_rynek', 'Wrocław_laczny_rynek'])

df_melt.columns = ['rok', 'miasto', 'wartość']

# Perform Repeated Measures ANOVA
aovrm = AnovaRM(df_melt, 'wartość', 'miasto', within=['rok'])
res = aovrm.fit()

print(res)
```
