```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import mixedlm
import numpy as np
```

<h2> Dane 08.2023 - 04.2024 <h2>

```python
domy_credyt_all_filtered = pd.read_csv('../Datasets/Price_During_Credit(08.2023-03.2024)/Ready_to_use/domy_credyt_all_GOOD_filtered_cols.csv')
domy_credyt_all_filtered.head()
```

```python
domy_credyt_all_filtered = domy_credyt_all_filtered.replace([np.inf, -np.inf], np.nan).dropna()\
    
y = domy_credyt_all_filtered['price']

X = domy_credyt_all_filtered[['squareMeters', 'rooms', 'buildYear']]
X = sm.add_constant(X)  

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
domy_credyt_all_filtered['squareMeters'] = domy_credyt_all_filtered['squareMeters'].astype(float)
domy_credyt_all_filtered['rooms'] = domy_credyt_all_filtered['rooms'].astype(float)
domy_credyt_all_filtered['floor'] = domy_credyt_all_filtered['floor'].astype(float)
domy_credyt_all_filtered['buildYear'] = domy_credyt_all_filtered['buildYear'].astype(float)
domy_credyt_all_filtered['centreDistance'] = domy_credyt_all_filtered['centreDistance'].astype(float)
domy_credyt_all_filtered['price'] = domy_credyt_all_filtered['price'].astype(float)

# Filling in the missing values in the ‘buildYear’ column with the median
domy_credyt_all_filtered['buildYear'].fillna(domy_credyt_all_filtered['buildYear'].median(), inplace=True)

#(One-Hot Encoding)
df = pd.get_dummies(domy_credyt_all_filtered, columns=['type', 'ownership', 'hasParkingSpace', 'hasBalcony', 'hasStorageRoom'], drop_first=True)

X = domy_credyt_all_filtered.drop(columns=['city', 'condition', 'price', 'date'])
y = df['price']

#(intercept)
X = sm.add_constant(X)

# Carrying out regression
model = sm.OLS(y, X).fit()

print(model.summary()) 
#???
```

```python
df_all_2010_2023_wybrane_kolumy_miasta = df_all_2010_2023_wybrane_kolumy_miasta.replace([np.inf, -np.inf], np.nan).dropna()\

y = df_all_2010_2023_wybrane_kolumy_miasta[['Gdańsk_laczny_rynek', 'Kraków_laczny_rynek', 'Poznań_laczny_rynek','Warszawa_laczny_rynek', 'Wrocław_laczny_rynek']]

X = df_all_2010_2023_wybrane_kolumy_miasta['rok']
X = sm.add_constant(X)  

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python

cities = ['Gdańsk_laczny_rynek', 'Kraków_laczny_rynek', 'Poznań_laczny_rynek', 
          'Warszawa_laczny_rynek', 'Wrocław_laczny_rynek']

for city in cities:
    y = df_all_2010_2023_wybrane_kolumy_miasta[city]
    X = df_all_2010_2023_wybrane_kolumy_miasta[['rok']]
    X = sm.add_constant(X)  
    model = sm.OLS(y, X).fit()
    print(f"Regresja dla {city}:")
    print(model.summary())
    print("\n")
```
